```python
import pandas as pd
!pip install wikipedia
import wikipedia as wp
 
#Get the html source
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")
df = pd.read_html(html)[0]

#assigning first row as a header
df_header = df.iloc[0]
df = df[1:] 
df.columns = df_header 


df_b = df[df['Borough'] != 'Not assigned']

df_gb = df_b.groupby(by=['Postcode','Borough'],sort=False).agg(lambda x: ','.join(x))

df_gb.reset_index(level=['Postcode','Borough'], inplace=True)
```

```python
df_geo = pd.read_csv('https://cocl.us/Geospatial_data')
df_geo.columns = ['Postcode','Latitude','Longitude']
df_geo
```

```python
#df_gb.set_index('Postcode').join(df_geo.set_index('Postal code'))
df_merge = pd.merge(df_gb, df_geo, on='Postcode', how='inner')
df_merge
```

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_t = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, df_merge in zip(df_merge['Latitude'], df_merge['Longitude'], df_merge['Borough'], df_merge['Neighbourhood']):
    label = '{}, {}'.format(df_merge, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_t)  
    
map_t
```
